# Loss Functions

## In context...
"Loss" refers to a measure of the error or discrepancy between the predicted output of a machine learning model and the actual or desired output. 


Common loss functions for different machine learning tasks:
- Mean Squared error, used in regression problems 
- Binary Cross-Entropy, used in classification
- Categorical Cross-Entropy, used in multi-class classification problems 

## What Loss Function does ChatGPT use?
ChatGPT uses maximum likelihood estimation (MLE) as an objective function, rather than a specific loss function. 

During training, the model is trained to predict the next word in a sequence given the previous words. The goal is to maximize the likelihood of generating the correct next word. 

#### The negative log-likelihood loss function, often referred to as cross-entropy loss: 
**NLL = -log(P(true next word | previous words))** 

This loss function penalizes the model more when it assigns low probabilities to the correct next word, encouraging it to learn to generate more accurate and coherent responses.
